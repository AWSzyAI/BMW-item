```python
import pandas as pd
m_78 = pd.read_csv('../data/m_78.csv')
m_78.head()
```

```python
m_78["itemcreationdate"].unique()
```

```python
# m_78["itemcreationdate"]格式化为data类型
m_78['itemcreationdate'] = pd.to_datetime(m_78['itemcreationdate'])
m_78.head()
```

```python

# 根据itemcreationdate划分数据得到m_7和m_8
m_7 = m_78[m_78['itemcreationdate'] < '2025-08-01']
m_8 = m_78[m_78['itemcreationdate'] >= '2025-08-01']
```

```python
m_7.head()
```

```python
m_8.head()
```

```python
m_7.to_csv('../data/m_7.csv', index=False)
m_8.to_csv('../data/m_8.csv', index=False)
```

```python
# 合并m_34,m_56,m_7为训练集m_train,m_8为测试集m_test
m_34 = pd.read_csv('../data/m_34.csv')
m_56 = pd.read_csv('../data/m_56.csv')
m_train = pd.concat([m_34, m_56, m_7])
m_test = m_8
m_train.to_csv('../data/m_train.csv', index=False)
m_test.to_csv('../data/m_test.csv', index=False)
```
