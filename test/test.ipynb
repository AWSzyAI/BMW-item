```python
import pandas as pd
df = pd.read_csv("../data/case_item_1021(in).csv")
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.columns
```

```python
# 检查一下所有字段一模一样的重复的行，确认没有重复的
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
# 按照case_submitted_date 转换成datetime格式
df['case_submitted_date'] = pd.to_datetime(df['case_submitted_date'])
df.head()
```

```python
# 发现有的样本有case_id	case_title	performed_work	case_submitted_date，但是没有extern_id	linked_items	itemcreationdate	item_title	
# 单独整理成一个未知item的cases的数据集，并将原来数据集中的这些样本删除
unknown_item_df = df[df['linked_items'].isnull()]
unknown_item_df.to_csv("../output/unknown_item_cases.csv", index=False)
df = df.drop(unknown_item_df.index)
df
```

```python
df.shape, unknown_item_df.shape
```

```python
# 统计不同月份的样本数量
df['month'] = df['case_submitted_date'].dt.to_period('M')
monthly_counts = df['month'].value_counts().sort_index()
monthly_counts
```

```python
df
```

```python
# 以34567月为train和eval，8月的数据为test，划分数据集
df['month'] = df['case_submitted_date'].dt.strftime('%Y-%m')
train_eval_df = df[df['month'].isin(['2025-03', '2025-04', '2025-05', '2025-06', '2025-07'])]
test_df = df[df['month'] == '2025-08']

print(train_eval_df.shape, test_df.shape)

train_eval_df.to_csv("../output/m_train_eval.csv", index=False)
test_df.to_csv("../output/m_test.csv", index=False)
```
